```python
from scipy.sparse import *
import pandas as pd
import numpy as np
from sample import sample_generator
```

```python
rng = sample_generator(3, 1000)
```

```python
P = 2**rng.p
```

```python
rows_dict = {i: np.argwhere(rng.barcode.reshape(-1) == i).reshape(-1) for i in range(P)}
ones = {i: np.array([1 for _ in range(len(arr))]) for i,arr in rows_dict.items()}
N = rng.X.shape[0]
```

```python
ones[0]
```

```python
rows_dict[0]
```

```python
def gen_crs_arr(val):
    rows = rows_dict[val]
    one = ones[val]
    array = csc_matrix((one, ([0 for _ in range(len(rows))], rows)), shape = (1, N))
    return array
```

```python
def gen_connectivity_matrix(barcode):
    crs_arrays = []
    for b in rng.barcode.reshape(-1):
        array = gen_crs_arr(b)
        crs_arrays.append(array)
    return vstack(crs_arrays).tocoo()
```

```python
connectivity_matrix = gen_connectivity_matrix(rng.barcode)
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, connectivity= connectivity_matrix, linkage = 'average')
```

# Average Linkage

```python
cluster.fit(rng.barcode, rng.y.reshape(-1, 1))
```

```python
df = pd.DataFrame(zip(rng.barcode.reshape(-1), cluster.labels_), columns = ['full','reduced'])
df.groupby('full').agg(np.mean)
```
